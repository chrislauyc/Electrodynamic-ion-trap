```python
#Construct a surface of quadrupole and hexapole electric field
# reference: http://www.massspecpro.com/technology/ion-optics/multipole-ion-guide
import numpy as np
x = np.linspace(-5,5,25)
Xv, Yv = np.meshgrid(x,x)
Z_quad = (Xv**2-Yv**2)/2**2 #equation of the field
Z_hex = (Xv**3-3*Xv*Yv**2)/2**3 #equation of the field
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(Xv,Yv,Z_hex)
ax.set_title("hexapole field")
ax.set_xlabel("x direction")
ax.set_ylabel("y direction")
ax.set_zlabel("potential")
plt.show()
```

```python
#plot hexapole electric field
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(Xv,Yv,Z_quad)
ax.set_title("quadrupole field")
ax.set_xlabel("x direction")
ax.set_ylabel("y direction")
ax.set_zlabel("potential")
plt.show()
```
