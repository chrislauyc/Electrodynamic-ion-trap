```python
#Construct a surface of quadrupole and hexapole electric field
# reference: http://www.massspecpro.com/technology/ion-optics/multipole-ion-guide
import numpy as np
x = np.linspace(-5,5,25)
Xv, Yv = np.meshgrid(x,x)
Z_quad = (Xv**2+Yv**2)/2**2 #equation of the field
Z_hex = (Xv**3-3*Xv*Yv**2)/2**3 #equation of the field
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.offline as py
py.init_notebook_mode(connected=True)
#plot quadrupole electric field
data = [
    go.Surface(
        z=Z_quad
    )
]

layout = go.Layout(
    title='Quadrupole potential',
    autosize=True,
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='elevations-3d-surface')
```

```python
#plot hexapole electric field
data = [
    go.Surface(
        z=Z_hex
    )
]

layout = go.Layout(
    title='Hexapole potential',
    autosize=True,
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='elevations-3d-surface')
```
